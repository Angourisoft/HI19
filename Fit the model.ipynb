```python
from hypo2.addit.functions import Functional as F
paths = F.gen_paths("./data2")[:60]
```

```python
import random
from hypo2.api import FitEnv

def fitit(class_count):
    cfg = MyConfig()
    cfg.CLASS_COUNT = class_count
    cfg.N_EPOCHS = class_count * 300
    id = random.randint(0, len(paths) - cfg.CLASS_COUNT - 1)
    pp = paths[id : id + cfg.CLASS_COUNT]
    fit = FitEnv(cfg)
    fit.fit(pp, verbose=True, plot=True)
    fit.save()
    Cache(cfg).clear()
```
